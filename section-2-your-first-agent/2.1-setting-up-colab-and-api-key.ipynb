## Setting Up Your API Key with Colab Secrets

We store the API key in **Colab Secrets** — a secure panel that keeps credentials out of your notebook file entirely.

### Steps

1. Click the **key icon** in the left sidebar (it looks like a small padlock/key)
2. Click **+ Add new secret**
3. Set the **Name** to exactly: `ANTHROPIC_API_KEY`  
   *(capital letters, underscores, no spaces)*
4. Paste your Anthropic API key as the **Value**
5. Toggle **Notebook access** to **ON**
6. Close the Secrets panel and continue to the next cell

> **Why Colab Secrets?** If you put your API key directly in a code cell, it gets saved inside the notebook file. Anyone you share the notebook with can see it. Colab Secrets stores the key separately so your notebook stays safe to share.

```python
# Cell 4 — Load the API key from Colab Secrets into the environment
# The SDK will automatically pick up ANTHROPIC_API_KEY from os.environ.
# If you see a pop-up asking to grant notebook access, click Allow.

from google.colab import userdata
import os

os.environ["ANTHROPIC_API_KEY"] = userdata.get("ANTHROPIC_API_KEY")
```
