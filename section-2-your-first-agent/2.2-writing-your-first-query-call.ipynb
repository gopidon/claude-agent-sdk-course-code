## Writing Our First `query()` Call

The `query()` function is the heart of the Claude Agent SDK.

You give it:
- A **`prompt`** — what you want the agent to do
- A **`ClaudeAgentOptions`** object — configuration, including which tools the agent is allowed to use

It returns an **async stream of messages**. Each message represents something the agent did:
- A tool it decided to call
- The result of that tool call
- The final answer

You loop over that stream with `async for`.

We'll look at message types in detail in **Lecture 2.3**. For now, we'll print everything so you can see the full stream live.

```python
# Cell 4 — Our First query() Call
#
# We ask Claude what files are in the current directory.
# We give it two tools: Bash (to run shell commands) and Glob (to find files by pattern).
#
# Claude will autonomously decide which tool(s) to use and in what order.
# Watch the output — you will see tool calls, tool results, and a final answer.
#
# You didn't tell it HOW to answer. You gave it a goal and the tools to reach it.
# That is an agent.

from claude_agent_sdk import query, ClaudeAgentOptions

async for message in query(
    prompt="What files are in this directory?",
    options=ClaudeAgentOptions(allowed_tools=["Bash", "Glob"]),
):
    print(message)
```

## What Just Happened?

Let's connect the output back to what we know.

1. **Claude received the prompt and the list of allowed tools**  
   It entered its internal agent loop.

2. **Claude chose a tool**  
   It decided — on its own — to call `Bash` or `Glob` to inspect the file system.  
   You didn't tell it which one. That decision was Claude's.

3. **The tool ran and returned a result**  
   Claude read the output and decided whether it had enough information.

4. **Claude exited the loop and returned a final answer**  
   That final answer arrived as the last message in the stream — a `ResultMessage`.

5. **You looped over the stream with `async for`**
